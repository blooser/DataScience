```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('bank-full.csv', delimiter=';')
```

```python
dataset
```

```python
X = dataset.iloc[:, 0:16].values
y = dataset.iloc[:, 16].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

label_x = LabelEncoder()
label_y = LabelEncoder()
```

```python
dummy_indexes = [1, 2, 3, 4, 6, 7, 8, 10, 15]
for i in  dummy_indexes:
    X[:, i] = label_x.fit_transform(X[:, i])
y = label_y.fit_transform(y)

onehotencoder = OneHotEncoder(categorical_features=dummy_indexes)
X = onehotencoder.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def built_classifier():
    classifier = Sequential()
    
    classifier.add(Dense(48, input_dim=51, kernel_initializer="uniform", activation="relu"))
    classifier.add(Dropout(0.2))
    
    classifier.add(Dense(86, kernel_initializer="uniform", activation="relu"))
    classifier.add(Dropout(0.15))
    
    classifier.add(Dense(86, kernel_initializer="uniform", activation="relu"))
    classifier.add(Dropout(0.15))
    
    classifier.add(Dense(120, kernel_initializer="uniform", activation="softmax"))
    classifier.add(Dropout(0.13))
    
    classifier.add(Dense(1, kernel_initializer="uniform", activation="sigmoid"))
    
    classifier.compile("adam", "binary_crossentropy", ["accuracy"])
    
    return classifier
```

```python
classifier = KerasClassifier(build_fn=built_classifier, batch_size=14, epochs=25)
accuracy = cross_val_score(classifier, X_train, y_train, cv=10, n_jobs=10)
```

```python
def percent(value):
    return "{0:.0%}".format(value)
```

```python
accuracy
```

```python
for single_acc in accuracy:
    print("{}".format(percent(single_acc)), end=' ')
```

```python
print("Real Accuracy: {}".format(percent(accuracy.mean())))
print("Loss: {}".format(percent(accuracy.std())))
```
